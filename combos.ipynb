```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (4, 3)
plt.rcParams["figure.dpi"] = 300
```

```python
X = ['1', '2', '3', '4', '5', '6', '...']
Y = [10, 6, 6, 3, 2, 2, 2]

plt.plot(X, Y, color="gray", marker="o")
plt.title('Hanging task')
plt.xlabel('consequtive demonstrations')    
plt.ylabel('number of task rules')
```

```python
X = ['1', '2', '3', '4', '...']
Y = [7, 6, 2, 2, 2]

plt.plot(X, Y, color="gray", marker="o")
plt.title('Stacking task')
plt.xlabel('consequtive demonstrations')    
plt.ylabel('number of task rules')
```
